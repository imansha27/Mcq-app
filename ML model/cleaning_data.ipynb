```python
import numpy as np
import pandas as pd 
import seaborn as sns 

MCQ=pd.read_csv('data.csv')
MCQ.head(10)
```

```python
#checking for missing values
MCQ.isnull().sum()
```

```python
# Getting rows with at least one null value
rows_with_null_values = MCQ[MCQ.isnull().any(axis=1)]

# Displaying the rows with null values
print("Rows with null values:")
print(rows_with_null_values)
```

```python
#drop the null rows
MCQ_cleaned = MCQ.dropna()
MCQ.dropna(inplace=True)
```
